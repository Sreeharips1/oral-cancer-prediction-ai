```python
import pandas as pd

df = pd.read_csv("../dataset/final_ml_dataset.csv")

df.head()
```

```python
print("Dataset Shape:")
print(df.shape)

print("\nColumns:")
print(df.columns[:10])
```

```python
df["Label"].value_counts()
```

```python
import matplotlib.pyplot as plt

df["Label"].value_counts().plot(
    kind="bar"
)

plt.title(
    "Healthy vs Oral Cancer Samples"
)

plt.xticks(
    [0,1],
    ["Healthy","Cancer"],
    rotation=0
)

plt.ylabel("Number of Samples")

plt.show()
```

```python
print(
    "Missing Values:",
    df.isnull().sum().sum()
)
```
