```python
import pandas as pd
import joblib

df = pd.read_csv(
    "../dataset/final_selected_dataset.csv"
)

model = joblib.load(
    "../backend/models/oral_cancer_rf.pkl"
)
```

```python
X = df.drop(
    "Label",
    axis=1
)

feature_importance = pd.DataFrame(
    {
        "Gene": X.columns,
        "Importance":
        model.feature_importances_
    }
)

feature_importance = (
    feature_importance
    .sort_values(
        by="Importance",
        ascending=False
    )
)
```

```python
feature_importance.head(20)
```

```python
import matplotlib.pyplot as plt

top20 = (
    feature_importance
    .head(20)
)

plt.figure(figsize=(10,6))

plt.barh(
    top20["Gene"],
    top20["Importance"]
)

plt.title(
    "Top 20 Important Genes"
)

plt.xlabel(
    "Importance Score"
)

plt.tight_layout()

plt.show()
```
